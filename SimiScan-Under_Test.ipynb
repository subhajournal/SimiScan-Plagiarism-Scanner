```python
def checkPlag(qr,lvl):
    perplag=[]
    plaglink=[]
    #type(query)
    query=qr
    #query = bytes(qr, 'utf-8') 
    
    try: 
        from googlesearch import search 
    except ImportError:  
        print("No module named 'google' found") 

    # to search 
    # query = "Each base classifier is trained in parallel with a training set which is generated by randomly drawing, with replacement, N examples(or data) from the original training dataset – where N is the size of the original training set. Training set for each of the base classifiers is independent of each other. Many of the original data may be repeated in the resulting training set while others may be left out."

    findlist=[]
    
    try:
        for j in search(query, tld="com", num=5, stop=5, pause=2): 
            findlist.append(j)
    except:
        messagebox.showerror('Connection Error!!', 'Server Overloaded')
        #print(findlist)
    dict_counter=[]
    for lnks in range(len(findlist)):
        url=findlist[lnks]   # Link Index
        content_str=""
        out=""
        spl=["//","\n","©","</",">","[document]","noscript","header","html",'meta','head', 'input','script',"•"]
        #open with GET method 
        resp=requests.get(url) 

        #http_respone 200 means OK status 
        if resp.status_code==200: 
                # we need a parser,Python built-in HTML parser is enough . 
                soup=BeautifulSoup(resp.text,'lxml')    
                text = soup.find_all(text=True)
                blacklist = [
                    '[document]',
                    'noscript',
                    'header',
                    'html',
                    'meta',
                    'head', 
                    'input',
                    'script',
                    # there may be more elements you don't want, such as "style", etc.
                ]

                for t in text:
                    if t.parent.name not in blacklist:
                        content_str += '{} '.format(t)
                for i in content_str:
                    if i in spl:
                        continue
                    else:
                        out+=i
        else: 
                print("Error") 

        #print(len(query.split()))
        formed=[]

        spt=0
        ept=lvl
        
        cut=ept-1
        for i in query.split():
            if query.split()[-1] not in formed:
                formed.append(query.split()[spt:ept])
                spt+=1
                ept+=1
        formed=formed[:-cut]
        #print(formed)
        arr=np.array(formed)
        #print(arr)

        finlist=[]

        #Joining Strings from List Elements
        s = " ";

        # This is sequence of strings.
        for i in range(len(formed)):
            finlist.append(s.join( formed[i] ))
        #print(finlist)

        dct1={finlist[i]:out.count(finlist[i])
             for i in range(len(finlist))
            }
        #print(dct1)
        ##################dict_counter.append(dct1)
        
        temparr=arr.flatten()
        perplag.append(int(sum(dct1.values())/len(temparr)*100))
        plaglink.append(findlist[lnks])
    #print("Plagiarised Words: ",sum(dct1.values()))     #To be implemented
    
    highplag=[]
    df=pd.DataFrame(columns=('Link','Plag(%)'))
    for i in range(len(findlist)):
        if perplag[i]!=0:
            print("Percentage of Plagiarism: ",perplag[i],"%")
            print("Links: ",findlist[i])
            highplag.append(perplag[i])
            d=pd.DataFrame([{'Link':findlist[i],'Plag(%)':perplag[i]}])
            df.append(d,ignore_index = True)
    df.to_csv("D:/Plagiarism_Result.csv")
            
    #print("Plagiarism: ",np.mean(np.array(highplag)),"%")
    #print("Total Matched Words: ",int(int(len(query.split()))/int(np.median(np.array(highplag)))))
    
    highplag=[]
    plaglinks=[]
    
    if len(findlist)>0:
        for i in range(len(findlist)):
            if perplag[i]!=0:
                #print("Percentage of Plagiarism: ",perplag[i],"%")
                #print("Links: ",findlist[i])
                plaglinks.append(findlist[i])
                highplag.append(perplag[i])
        df=pd.DataFrame({"Link":plaglinks,"Plag":highplag})
        #print("Plagiarism: ",int(np.mean(np.array(highplag))),"%")
        #print("Total Matched Words: ",int(int(len(query.split()))/int(np.median(np.array(highplag)))))
        #print(df.head(len(df)))
        time=str(datetime.datetime.now())
        time1=time.split()
        time1=''.join(time1)
        df.to_html("SimiScan Report.html")
        if len(highplag)>0:
            return np.mean(np.array(highplag))
        else:
            return 0
    else:
        return 0
    
    
```
